```python
#LINEAR REGRESSION WITH ONE VARIABLE (Predicting Sales through TV advertisements)
#GRADIENT DESCENT

#Importing packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)
```

```python
#Preprocessing the data
data=pd.read_csv('TVandSales.csv')
data
X=data.iloc[:,1]
Y=data.iloc[:,2]
plt.scatter(X,Y)
plt.xlabel('TV Advertisements')
plt.ylabel('Sales')
plt.show()
```

```python
#BUILDING THE MODEL
m=0
b=0
L=0.00001
epochs=50
n=float(len(X))
for i in range (epochs):
    Y_predict=m*X+b
    PartialDer_m=(-2/n) * sum(X*(Y-Y_predict))
    PartialDer_b=(-2/n) * sum(Y-Y_predict)
    m=m-L*PartialDer_m
    b=b-L*PartialDer_b
    
print(m,b)
```

```python
#MAKING PREDICTIONS
Y_predict=m*X+b
plt.scatter(X,Y)
plt.plot([min(X),max(X)],[min(Y_predict),max(Y_predict)],color='orange')
plt.show()
```

```python
import statsmodels.formula.api as smf
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataxx = pd.read_csv('TVandSales.csv')
dataxx
```

```python
dataxx.info()
```

```python
## STATSMODELS ###

# create a fitted model
lm1 = smf.ols(formula='Y ~ X', data=dataxx).fit()

# print the coefficients
lm1.params
```

```python
### SCIKIT-LEARN ###

# create X and y
TVads = ['X']
X = dataxx['TV'].values.reshape(-1,1)
y = dataxx.Sales

# instantiate and fit
lm2 = LinearRegression()
lm2.fit(X, y)

# print the coefficients
print(lm2.intercept_)
print(lm2.coef_)
```

```python
sns.pairplot(dataxx)
```

```python
# Preprocessing Input data
XX = dataxx.iloc[:, 1]
YY = dataxx.iloc[:, 2]
plt.xlabel('TV Advertisements')
plt.ylabel('Sales')
plt.scatter(XX, YY)
plt.show()
```

```python
Y_predx = lm2.coef_*XX +  lm2.intercept_
#Y_predx = 0.04753664*XX + 7.032593549127693

plt.scatter(XX, YY) 
plt.plot([min(XX), max(XX)], [min(Y_predx), max(Y_predx)], color='red')  # regression line
plt.show()
```

```python
#Y_predx = lm2.coef_*XX +  lm2.intercept_
#0.04753664, 7.032593549127693
Y_predx = 0.04753664*XX + 7.032593549127693

plt.scatter(XX, YY) 
plt.plot([min(XX), max(XX)], [min(Y_predx), max(Y_predx)], color='red')  # regression line
plt.show()
```

```python
m = 0
c = 0


L = 0.00001  # The learning Rate
epochs = 50  # The number of iterations to perform gradient descent

n = float(len(XX)) # Number of elements in X

# Performing Gradient Descent 
for i in range(epochs): 
    Y_pred = m*XX + c  # The current predicted value of Y
    D_m = (-2/n) * sum(XX * (YY - Y_pred))  # Derivative wrt m
    D_c = (-2/n) * sum(YY - Y_pred)  # Derivative wrt c
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update c

print (m, c)
```

```python
Y_pred = m*XX + c

plt.scatter(XX, YY) 
plt.plot([min(XX), max(XX)], [min(Y_pred), max(Y_pred)], color='red')  # regression line
plt.show()
```
